```python
import re
import nltk
import string
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt


from wordcloud import WordCloud
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from wordcloud import WordCloud, STOPWORDS
from nltk.tokenize import RegexpTokenizer, word_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
komen1 = open('/content/artikel.txt', 'r')

komentar1 = komen1.read()
```

```python
print(komentar1 )
```

```python
komentar1 = komentar1.lower()
print(komentar1)
```

```python
komentar1 = re.sub(r"\d+", "", komentar1)
print(komentar1)
```

```python
komentar1 = komentar1.translate(str.maketrans("","",string.punctuation))
print(komentar1)
```

```python
print(komentar1)
```

```python
nltk.download('popular')
```

```python
word_tokens = word_tokenize(komentar1)
```

```python
print(word_tokens)
```

```python
stop_words = set(stopwords.words('indonesian'))

word_tokens_no_stopwords = [w for w in word_tokens if not w in stop_words]
```

```python
print(word_tokens_no_stopwords)
```

```python
freq_kata_1 = nltk.FreqDist(word_tokens)
print(freq_kata_1.most_common(20))
freq_kata_1.plot(20)
plt.show()
```

```python
print(freq_kata_1.most_common(20))
```

```python
freq_kata_2 = nltk.FreqDist(word_tokens_no_stopwords)
print(freq_kata_2.most_common(20))
freq_kata_2.plot(20)
plt.show()
```

```python
print(freq_kata_2.most_common(20))
```
